Importing Libraries

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

Data processing

```python
dataset = pd.read_csv('Market_Basket_Optimisation.csv', header = None)
```

```python
transaction_t = []
for i in range(0, 7500):
    transaction_t.append([str(dataset.values[i,j]) for j in range(0,20)])
    
    
```

Training Apriori model on dataset

```python
from apyori import apriori
rules = apriori(transactions= transaction_t, min_support = 0.003, min_confidence = 0.2, min_lift = 3, min_lenght = 2, max_lenght = 2)
```

Visualising the model

```python
results = list(rules)
print(results)
```

Putting results in Pandas Dataframe for better visualisation

```python
def inspect(results):
    lhs = [tuple(result[2][0][0])[0] for result in results]
    rhs = [tuple(result[2][0][1])[0] for result in results]
    support = [result[1] for result in results]
    confidence =  [result[2][0][2] for result in results]
    lifts = [result[2][0][3] for result in results]
    
    return list(zip(lhs, rhs, support, confidence, lifts))


resultsindataframe = pd.DataFrame(inspect(results), columns =['lhs', 'rhs', 'support', 'confidence', 'lifts'])
```
