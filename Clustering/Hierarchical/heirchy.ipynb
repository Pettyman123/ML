** H CLUSTERING **

Importing librabries

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

Importing dataset

```python
dataset = pd.read_csv('Mall_Customers.csv')
X = dataset.iloc[:, [3,4]].values
```

Using Dendogram to find the optional number fo clusters

```python
import scipy.cluster.hierarchy as dhc
dendogram = dhc.dendrogram(dhc.linkage(X, method='ward'))
plt.title("Dendogram")
plt.xlabel('Customers')
plt.ylabel('Euclidian distances')
```

Training the HC model on dataset

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters = 5, linkage = 'ward')
y_hc = hc.fit_predict(X)
                                                    
```

```python
print(y_hc)
```

Visualising the cluster

```python
plt.scatter(X[y_hc==0, 0], X[y_hc==0, 1], s=50, color='red', label='Cluster 1')
plt.scatter(X[y_hc==1, 0], X[y_hc==1, 1], s=50, color='blue', label='Cluster 2')
plt.scatter(X[y_hc==2, 0], X[y_hc==2, 1], s=50, color='green', label='Cluster 3')
plt.scatter(X[y_hc==3, 0], X[y_hc==3, 1], s=50, color='pink', label='Cluster 4')
plt.scatter(X[y_hc==4, 0], X[y_hc==4, 1], s=50, color='cyan', label='Cluster 5')

plt.title("HC Clustering")
plt.xlabel('Anual income')
plt.ylabel('Spending score')
plt.legend()
plt.show()
```
