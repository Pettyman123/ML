** K MEANS CLUSTERING **

Importing librabries

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

Importing dataset

```python
dataset = pd.read_csv('Mall_Customers.csv')
X = dataset.iloc[:, [3,4]].values
```

Using elbow method to find the optional number fo clusters

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    k_means = KMeans(n_clusters = i, init='k-means++', random_state = 42)
    k_means.fit(X)
    wcss.append(k_means.inertia_)
    
```

```python
plt.plot(range(1,11), wcss, color='red')
plt.title("The elbow method")
plt.xlabel("Number of cluster")
plt.ylabel("WCSS")
plt.show()
```

Training the Kmeans model on dataset

```python
k_means = KMeans(n_clusters=5, init='k-means++', random_state=42)
y_k_means = k_means.fit_predict(X)
```

```python
print(y_k_means)
```

Visualising the cluster

```python
plt.scatter(X[y_k_means==0, 0], X[y_k_means==0, 1], s=50, color='red', label='cluster 1')
plt.scatter(X[y_k_means==1, 0], X[y_k_means==1, 1], s=50, color='blue', label='cluster 2')
plt.scatter(X[y_k_means==2, 0], X[y_k_means==2, 1], s=50, color='green', label='cluster 3')
plt.scatter(X[y_k_means==3, 0], X[y_k_means==3, 1], s=50, color='yellow', label='cluster 4')
plt.scatter(X[y_k_means==4, 0], X[y_k_means==4, 1], s=50, color='pink', label='cluster 5')
plt.scatter(k_means.cluster_centers_[:,0], k_means.cluster_centers_[:, 1], s = 100, label='Centroid', color='purple')
plt.xlabel("Number of cluster")
plt.ylabel("WCSS")
plt.legend()
plt.show()


```
